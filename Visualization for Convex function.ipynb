```python
def f(x):
    return (0.4*x[0]**2 + 0.4*x[1]**2)
```

```python
def fo():
    x = -1.5
    y = -1.5
    #z = np.random.random() - 3
    af = []
    qx = []
    qy = []
    for i in range(600):
        f = (0.4*x**2 + 0.4*y**2)
        x = x - 0.01 * (0.8 * x) 
        y = y - 0.01 * (0.8 * y) 
        qx.append(x)
        qy.append(y)
        af.append(f)
    return qx, qy, af
```

```python
def fm():
    x = -1.5
    y = -1.5
    pr_x = 0
    pr_y = 0
    gamma = 0.93
    #z = np.random.random() - 3
    af1 = []
    qx1 = []
    qy1 = []
    for i in range(600):
        f = (0.4*x**2 + 0.4*y**2)
        xx = (gamma * pr_x) + 0.01 * (0.8 * x) 
        yy = (gamma * pr_y) + 0.01 * (0.8 * y) 
        x = x - xx
        y = y - yy
        pr_x = xx
        pr_y = yy
        qx1.append(x)
        qy1.append(y)
        af1.append(f)
    return qx1, qy1, af1
```

```python
def fa():
    x = -1.5
    y = -1.5
    pr_x = 0
    pr_y = 0
    gamma = 0.93
    af1 = []
    qx1 = []
    qy1 = []
    for i in range(600):
        v_x = gamma * pr_x
        v_y = gamma * pr_y
        f =  (0.4*x**2 + 0.4*y**2)
        xx =  (0.8 * (x  - v_x)) 
        yy =  (0.8 * (y - v_y)) 
        v_x = gamma * pr_x + 0.01 * xx
        v_y = gamma * pr_y + 0.01 * yy
        x = x - v_x
        y = y - v_y
        pr_x = v_x
        pr_y = v_y
        qx1.append(x)
        qy1.append(y)
        af1.append(f)
    return qx1, qy1, af1
```
